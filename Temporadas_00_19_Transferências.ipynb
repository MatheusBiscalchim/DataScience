```python
fut[fut['Transfer_fee'] == fut['Transfer_fee'].min()]
```

```python
times = dict()
aux=0

for i in fut['Team_from'].tolist(): 
  if i not in times:
    times[i] = fut['Transfer_fee'][aux]
    aux=aux+1
  else:
    times[i] = times[i] + fut['Transfer_fee'][aux]
    aux=aux+1

time = max(times, key=times.get)
vendas = max(times.values())
time_min = min(times, key=times.get)
vendas_min = min(times.values())
print("O time com mais vendas é o",time,"com",vendas)
print("O time com menos vendas é o",time_min,"com",vendas_min)
```
